```python
import os
import jieba
import random
from sklearn.naive_bayes import MultinomialNB
import matplotlib.pyplot as  plt
```

```python
def  textprocessing(folder_path,test_size=0.2):
    folder_list=os.listdir(folder_path)
    data_list=[]
    class_list=[]
    for folder in folder_list:
        new_foldee_path=os.path.join(folder_path,folder)
        files=os.listdir(new_foldee_path)
        j=1
        for file in files:
            if j>100:
                break
            with open(os.path.join(new_foldee_path,file),encoding='utf-8') as f:
                raw=f.read()
            word_cut=jieba.cut(raw)
            data_list.append(list(word_cut))
            class_list.append(folder)
            j+=1
    data_class_list=list(zip(data_list,class_list))#zip压缩合并，讲数据集与标签对应压缩
    random.shuffle(data_class_list)
    index=int(len(data_class_list)*test_size)+1
    train_list=data_class_list[:index]
    test_list=data_class_list[index:]
    train_data_list,train_class_list=list(zip(*train_list))
    test_data_list,test_class_list=zip(*test_list)
    all_words_dict={}
    for word_list in train_data_list:
        for word in word_list:
            if word in all_words_dict:
                all_words_dict[word]+=1
            else:
                all_words_dict[word]=1
    all_word_tuple_list=sorted(all_words_dict.items(),key=lambda f:f[1],reverse=True)
    all_word_list,all_word_nums=zip(*all_word_tuple_list)
    all_word_list=list(all_word_list)
    return all_word_list,train_data_list,test_data_list,train_class_list,test_class_list
    
        
```

```python
if __name__=='__main__':
    folder_path='./Sample/'
    all_word_list,train_class_list,test_data_list,train_class_list,test_class_list=textprocessing(folder_path)
    print(all_word_list)
```

```python
def makewordset(word_file):
    word_set=set()
    with open(word_file,'r',encoding='utf-8') as f:
        for line in f.readlines():
            word=line.strip()
            if len(word)>0:
                word_set.add(word)
    return word_set
```

```python
def word_dict(all_word_list,deleten,stopwords_set=set()):
    feature_word=[]
    n=1
    for t in range(deleten,len(all_word_list)):
        if n>1000:
            break
        if not all_word_list[t].isdigit() and all_word_list[t] not in stopwords_set \
        and 1<len(all_word_list[t])<5:
            feature_word.append(all_word_list[t])
        n+=1
    return feature_word
```

```python
def textfeature(train_data_list,test_data_list,feature_words):
    def text_feature(text,feature_words):
        text_words=set(text)
        features=[1 if word in text_words else 0 for word in feature_word]
        return features
    train_feature_list=[text_feature(text,feature_words)for text in train_data_list]
    test_feature_list=[text_feature(text,feature_words) for text in test_data_list]
    return train_feature_list,test_feature_list
```

```python
def textclassifier(train_feature_list,test_feature_list,train_class_list,test_class_list):
    test_accuracy=MultinomialNB().fit(train_feature_list,train_class_list).score(test_feature_list,test_class_list)
#     test_accuracy=classifier.score(test_feature_list,test_class_list)
    return test_accuracy
```

```python
if __name__=='__main__':
    stopwords_file='stopwords_cn.txt'
    stopwords_set=makewordset(stopwords_file)
    test_accuracy_list=[]
    deletens=range(0,1000,20)
    all_word_list,train_data_list,test_data_list,train_class_list,test_class_list=textprocessing(folder_path, test_size=0.2)
    feature_words=word_dict(all_word_list,10,stopwords_set)
    train_feature_list,test_feature_list=textfeature(train_data_list,test_data_list,feature_words)
    test_acccucy=textclassifier(train_feature_list,test_feature_list,train_class_list,test_class_list)
    test_accuracy_list.append(test_acccucy)
    print(test_acccucy)
```
